```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("heart.csv")

# Show first 5 rows
print("ðŸ”¹ Head of the Data:")
print(data.head())

# Shape of data
print("\nðŸ”¹ Shape of Data (rows, columns):")
print(data.shape)

# Show column names
print("\nðŸ”¹ Column Names:")
print(data.columns.tolist())

# Check for null values
print("\nðŸ”¹ Null Values per Column:")
print(data.isnull().sum())

# Show value counts for 'sex' column
print("\nðŸ”¹ Value Counts for 'sex' Column (1 = Male, 0 = Female):")
print(data['sex'].value_counts())

# âœ… Boxplot: 'trestbps'
sns.boxplot(x=data['trestbps'])
plt.title("trestbps checking")
plt.show()

filtered_trestbps = data[data['trestbps'] <= 170]
sns.boxplot(x=filtered_trestbps['trestbps'])
plt.title("Box Plot: trestbps (â‰¤ 170)")
plt.show()

# âœ… Boxplot: 'age'
plt.boxplot(x=data['age'])
plt.show()

# âœ… Bar Plot: Sex vs Max Heart Rate (thalach) â€” no warnings
plt.figure(figsize=(8, 5))
sns.barplot(x='sex', y='thalach', data=data, errorbar=None, hue='sex', palette="Set2", legend=False)
plt.title("Bar Plot: Sex vs Max Heart Rate (thalach)")
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Max Heart Rate (thalach)")
plt.show()
```
